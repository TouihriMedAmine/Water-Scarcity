```python
import os
from PIL import Image
from torch import torch
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader
import torch.optim as optim
import torch.nn as nn
import torch.nn.functional as F
from collections import defaultdict
import matplotlib.pyplot as plt
from einops import rearrange, repeat
from torch import einsum
import numpy as np

# ==== CONFIG ====
seq_len = 4   # Number of input frames (images) in the sequence
target_len = 4   # Number of target frames (images) to predict
image_size = (128, 128)  # Size to resize images to (height, width)
folder_path = '/kaggle/input/dataimages/data'  # Directory where images are stored
epochs = 30
batch_size = 8
learning_rate = 1e-4
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')


# ==== TRANSFORMS ====
transform = transforms.Compose([
    transforms.Resize(image_size),
    transforms.Grayscale(num_output_channels=1),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,)),
])

# ==== VIT MODEL ====
class VisionTransformer(nn.Module):
    def __init__(self, input_len=8, target_len=8, image_size=(64,64), dim=256, num_heads=4, num_layers=6):
        super().__init__()
        self.input_len = input_len
        self.target_len = target_len
        self.image_size = image_size
        self.dim = dim
        H, W = image_size
        # Add residual connections
        self.residual = nn.ModuleList([
            nn.Linear(dim, dim) for _ in range(num_layers)
        ])
        
        # Add multi-scale processing
        self.downsample = nn.Conv2d(dim, dim//2, kernel_size=3, stride=2, padding=1)
        self.upsample = nn.ConvTranspose2d(dim//2, dim, kernel_size=3, stride=2, padding=1)
        
        self.patch_size = 16  # 32x32 image with 8x8 patches = 4x4 grid
        self.num_patches = (H // self.patch_size) * (W // self.patch_size)
        self.patch_dim = 1 * self.patch_size * self.patch_size

        # Patch embedding
        self.patch_embed = nn.Conv2d(1, dim, kernel_size=self.patch_size, stride=self.patch_size)
        
        # Positional encodings
        self.temporal_pos = nn.Parameter(0.02 * torch.randn(1, input_len, dim))
        self.spatial_pos = nn.Parameter(0.02 * torch.randn(1, self.num_patches, dim))

        self.dropout = nn.Dropout(0.3)
        
        # Transformer
        self.transformer = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(d_model=dim, nhead=num_heads, batch_first=True),
            num_layers=num_layers
        )
        
        # Prediction head
        self.head = nn.Sequential(
            nn.Linear(dim, dim * 4),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(dim * 4, self.patch_dim)
        )

    def forward(self, x):
        # x shape: (B, T, C, H, W)
        B, T, _, H, W = x.shape
        
        # Embed patches
        x = rearrange(x, 'b t c h w -> (b t) c h w')
        x = self.patch_embed(x)  # (B*T, D, H', W')
        x = rearrange(x, '(b t) d h w -> b t (h w) d', b=B, t=T)
        
        # Add positional encodings
        x = x + self.spatial_pos.unsqueeze(1)  # Add spatial pos
        x = x + self.temporal_pos.unsqueeze(2)  # Add temporal pos
        
        x = self.dropout(x)
        
        # Transformer expects (seq_len, batch, dim)
        x = rearrange(x, 'b t n d -> (b n) t d')
        x = self.transformer(x)
        x = rearrange(x, '(b n) t d -> b t n d', b=B, n=self.num_patches)
        
        # Predict patches for all future frames
        future_patches = self.head(x)  # (B, T, N, 3*p*p)
        
        # Reshape patches back to images
        # First reshape to (B, T, N, C, p, p)
        future_patches = future_patches.view(B, T, self.num_patches, 1, self.patch_size, self.patch_size)
        
        # Combine patches into images
        # Calculate grid dimensions
        grid_h = H // self.patch_size
        grid_w = W // self.patch_size
        
        # Reshape to (B, T, grid_h, grid_w, C, p, p)
        future_patches = future_patches.view(B, T, grid_h, grid_w, 1, self.patch_size, self.patch_size)
        
        # Permute and reshape to (B, T, C, H, W)
        future_images = future_patches.permute(0, 1, 4, 2, 5, 3, 6).contiguous()
        future_images = future_images.view(B, T, 1, H, W)
        # ===== FORCE NEAR-WHITE PIXELS TO PURE WHITE =====
        threshold = 0.91  # Value between -1 and 1 in normalized space
        # Create a mask where all channels are > threshold
        white_mask = (future_images > threshold).all(dim=2, keepdim=True)  # (B, T, 1, H, W)
        # Set those pixels to 1 (pure white)
        future_images = future_images.masked_fill(white_mask, 1.0)
        
        return future_images

# ==== DATASET PREPARATION ====
def extract_month(filename):
    basename = os.path.splitext(filename)[0]
    datecode = basename.split('_')[-1]
    return datecode[:6]

def mask_white_pixels(image_tensor, threshold=0.9):
    """
    Mask white pixels (or near-white) by setting them to black (or any other value you prefer).
    The threshold determines the cutoff for how close to white a pixel should be to be masked.
    """
    # Convert to numpy array to manipulate
    image_np = image_tensor.permute(1, 2, 0).numpy()  # Change to HWC format
    
    # Create a mask where all pixels close to white are set to True
    white_mask = np.all(image_np >= (threshold * 255), axis=-1)
    
    # Set white pixels to black (0) or any value you'd prefer (e.g., 0 for black)
    image_np[white_mask] = 0
    
    # Convert back to tensor
    return torch.tensor(image_np).permute(2, 0, 1)  # Convert back to CHW format
def load_and_group_by_month(folder_path):
    month_dict = defaultdict(list)
    for fname in sorted(os.listdir(folder_path)):
        if fname.endswith(".png"):
            try:
                month = extract_month(fname)
                img_path = os.path.join(folder_path, fname)
                img = Image.open(img_path).convert("L")
                tensor = transform(img)
                
                month_dict[month].append(tensor)
            except Exception as e:
                print(f"Error loading {fname}: {e}")

    monthly_tensors = []
    for month in sorted(month_dict.keys()):
        tensors = month_dict[month]
        if len(tensors) >= seq_len + target_len:
            stacked = torch.stack(tensors)
            monthly_tensors.append(stacked)
            print(f"Month {month}: {len(tensors)} images")
    return monthly_tensors

class SimplePrecipDataset(Dataset):
    def __init__(self, monthly_sequences, input_len=8, target_len=8):
        self.samples = []
        self.monthly_sequences = monthly_sequences
        self.input_len = input_len
        self.target_len = target_len

        for month_tensor in self.monthly_sequences:
            total_len = input_len + target_len
            if month_tensor.shape[0] < total_len:
                continue
            for i in range(month_tensor.shape[0] - total_len + 1):
                input_seq = month_tensor[i : i + input_len]
                target_seq = month_tensor[i + input_len : i + total_len]
                self.samples.append((input_seq, target_seq))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        input_seq, target_seq = self.samples[idx]
        if torch.equal(input_seq, target_seq):
            print(f"Input and target are the same at index {idx}")
        return input_seq, target_seq

# ==== VISUALIZATION ====
def visualize_sequence(tensor_seq, title="Sequence", max_frames=8):
    n = min(max_frames, tensor_seq.shape[0])
    fig, axes = plt.subplots(1, n, figsize=(15, 3))

    if n == 1:
        axes = [axes]

    for i in range(n):
        img = tensor_seq[i].permute(1, 2, 0).detach().numpy()
        axes[i].imshow(img)
        axes[i].axis('off')
        axes[i].set_title(f"T{i+1}")

    plt.suptitle(title)
    plt.tight_layout()
    plt.show()

# ==== TRAINING ====
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision.models import vgg16
from torchvision.transforms import Normalize

class VGGPerceptualLoss(nn.Module):
    def __init__(self, device):
        super(VGGPerceptualLoss, self).__init__()
        vgg = vgg16(pretrained=True).features[:16].eval().to(device)
        for param in vgg.parameters():
            param.requires_grad = False
        self.vgg = vgg
        self.device = device

    def forward(self, pred, target):
        # Rescale from [-1, 1] to [0, 1]
        pred = (pred + 1) / 2
        target = (target + 1) / 2
        
        pred = torch.clamp(pred, 0, 1)
        target = torch.clamp(target, 0, 1)

        # ðŸ› ï¸ Make sure input has 3 channels
        if pred.shape[1] == 1:
            pred = pred.repeat(1, 3, 1, 1)
        if target.shape[1] == 1:
            target = target.repeat(1, 3, 1, 1)
        
        # Now safe to pass through VGG
        pred_features = self.vgg(pred)
        target_features = self.vgg(target)

        return F.mse_loss(pred_features, target_features)



def hybrid_mse_l1_loss(pred, target, threshold=0.97, alpha=1e-2, beta=1e-2, lambda_masked=0.7):
    """
    Combines full loss with masked loss to balance structure and focus.
    """
    # Mask for non-white pixels (important)
    mask = (target < threshold).float()
    mask = mask  # You can try mask instead of mask ** 2

    # --- Masked loss ---
    masked_mse = (((pred - target) ** 2) * mask).sum() / (mask.sum() + 1e-8)

    # --- Full loss ---
    full_mse = ((pred - target) ** 2).mean()

    # Combine masked and full MSE
    mse_combined = lambda_masked * (full_mse + 0.4*masked_mse)

    total_loss = mse_combined 

    # Print each loss term for monitoring
    print(f"Masked MSE: {masked_mse.item()}, Full MSE: {full_mse.item()}, Total Loss: {total_loss.item()}")

    return total_loss

    
from torch.utils.data import DataLoader, Subset

# Parameters for splitting dataset (you can adjust these based on your dataset)
train_size = int(0.8 * len(dataset))  # 80% for training
test_size = len(dataset) - train_size  # 20% for testing

# Manually create indices for training and testing
train_indices = list(range(train_size))  # Indices for training data
test_indices = list(range(train_size, len(dataset)))  # Indices for testing data

# Create Subsets
train_subset = Subset(dataset, train_indices)
test_subset = Subset(dataset, test_indices)
def train_model(dataset, model):
    dataloader = DataLoader(train_subset, batch_size=batch_size, shuffle=False)
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs)

    perceptual_loss_fn = VGGPerceptualLoss(device)
    
    for epoch in range(epochs):
        model.train()
        total_loss = 0

        for inputs, targets in dataloader:
            inputs, targets = inputs.to(device), targets.to(device)

            optimizer.zero_grad()

            # Model already includes tanh â€” no need to wrap again
            outputs = model(inputs)

            # Plot histogram every few batches (optional)
            
            pixel_loss = hybrid_mse_l1_loss(outputs, targets)
            # Initialize the perceptual loss
            perceptual_loss = 0.0
            
            # Loop through all frames (except batch dimension)
            for t in range(outputs.shape[1]):  # Iterate over time steps (frames)
                pred_frame = outputs[:, t]  # Get the frame from the output sequence (B, 3, H, W)
                target_frame = targets[:, t]  # Get the frame from the target sequence (B, 3, H, W)
                
                perceptual_loss += perceptual_loss_fn(pred_frame, target_frame)  # Add perceptual loss for each frame
            
            # Normalize the perceptual loss by the number of frames
            perceptual_loss /= outputs.shape[1]
            loss = pixel_loss + 0.1 * perceptual_loss  # Tune alpha as needed
            

            loss.backward()
            optimizer.step()

            total_loss += loss.item()
        scheduler.step()
        avg_loss = total_loss / len(dataloader)
        visualize_sequence(outputs[0].cpu(), "Predicted Sequence")
        print(f'Epoch {epoch+1}/{epochs}, Loss: {avg_loss:.4f}')

    return model
# ==== MAIN EXECUTION ====
if __name__ == "__main__":
    # Load and prepare dataset
    print("Loading and preparing dataset...")
    monthly_tensors = load_and_group_by_month(folder_path)
    dataset = SimplePrecipDataset(monthly_tensors, input_len=seq_len, target_len=target_len)
    print(f"Total sequences: {len(dataset)}")
    torch.save(dataset, "prep.pt")

    # Initialize model
    print("Initializing model...")
    model = VisionTransformer(input_len=seq_len, target_len=target_len, image_size=image_size).to(device)

    # Train model
    print("Starting training...")
    model = train_model(dataset, model)
    torch.save(model.state_dict(), "vit_model.pth")

    # Sample inference
    print("Running inference on the test set...")
    test_dataloader = DataLoader(test_subset, batch_size=batch_size, shuffle=False)
    model.eval()  # Set model to evaluation mode
    total_loss = 0
    with torch.no_grad():  # Disable gradient computation during inference
        for batch_idx, (inputs, targets) in enumerate(test_dataloader):
            inputs, targets = inputs.to(device), targets.to(device)
            
            # Forward pass
            predictions = model(inputs)
            
            # Calculate loss (if applicable, depending on your model)
            loss = hybrid_mse_l1_loss(predictions, targets)  # Replace with your loss function
            total_loss += loss.item()
    
            # Optionally: You can print predictions vs targets or calculate any other metrics like RMSE, MAE, etc.
            # For example, you can visualize the first image in the batch:
            if batch_idx == 0:
                visualize_sequence(targets[0].cpu(), "Test Target")
                visualize_sequence(predictions[0].cpu(), "Predicted Output")
    
    # Average loss over all batches
    average_loss = total_loss / len(test_dataloader)
    print(f"Test set average loss: {average_loss}")

    # Visualize results
    #visualize_sequence(targets[0].cpu(), "Input Sequence")
    #visualize_sequence(prediction[0].cpu(), "Predicted Sequence")
```
